```python
def StripPunctuation(s):
    words=[]
    for word in s:
        exclude="?:!.,;.-_`~+=#()/\|][*' "
        word = ''.join(ch if ch not in exclude else " " for ch in word )
        words+=word.split()
    return words

# def Suggest(s):
#     d = enchant.Dict("en_US")
#     result=[]
#     for i,elt in enumerate(s):
#         try:
#             if not d.check(elt):
#                 result+=d.suggest(elt)[0].split()
#                 print(d.suggest(elt)[0].split(),elt)
#             else: result+=[elt]
#         except: result+=[elt]
#     return result
    

def prepareWords(review):
    stopWords = set(stopwords.words("english"))   
    s = re.sub(r'<.*/>', ' .',review)  
    s=re.sub("[^a-zA-Z]", " ", s) 
    word_list=StripPunctuation(nltk.word_tokenize(s.lower()))

            
    s=[WordNetLemmatizer().lemmatize(elt.lower())\
       for elt in word_list if not ((elt in stopWords) or (len(elt)<2)) ]
    return( " ".join( s )) 





    
```

```python
train['review_cleaned']=train['review'].map(prepareWords)
print('train done')
test['review_cleaned']=test['review'].map(prepareWords)
```

# naive bayes

```python


    

cross_val_coefs=[]
kf = cross_validation.KFold(train.shape[0], n_folds=myCV,shuffle=True)
for j,(train_idx, test_idx) in enumerate(kf):
    train_current=train.ix[train_idx]
    test_current=train.ix[test_idx]


    pos_words=train_current[train_current['sentiment']==1]['review_cleaned'].map(lambda x:x.split())
    pos=[FreqDist(pos_rev) for pos_rev in pos_words]

    neg_words=train_current[train_current['sentiment']==0]['review_cleaned'].map(lambda x:x.split())
    neg=[FreqDist(neg_rev) for neg_rev in neg_words]
    
    pos_words=test_current[test_current['sentiment']==1]['review_cleaned'].map(lambda x:x.split())
    pos_test=[FreqDist(pos_rev) for pos_rev in pos_words]

    neg_words=test_current[test_current['sentiment']==0]['review_cleaned'].map(lambda x:x.split())
    neg_test=[FreqDist(neg_rev) for neg_rev in neg_words]

    add_label = lambda lst, lab: [(x, lab) for x in lst]
    classif = nltk.NaiveBayesClassifier.train(add_label(pos, 'pos') + add_label(neg, 'neg'))


    l_pos = np.array(classif.classify_many(pos_test))
    l_neg = np.array(classif.classify_many(neg_test))
    print "validation score: %0.2f" % (
              ((l_pos == 'pos').sum()+(l_neg == 'neg').sum())/float(len(l_pos)+len(l_neg) ))







```
