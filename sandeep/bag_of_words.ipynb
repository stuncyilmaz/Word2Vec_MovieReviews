```python
import pandas as pd
import numpy as np
import nltk
import re
import json
import unicodedata
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
#import enchant
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import cross_validation
from nltk.probability import FreqDist
```

```python
train=pd.DataFrame.from_csv('labeledTrainData.tsv', sep='\t')
train['review']=train['review'].map(lambda x: x.decode("utf8") ).\
map(lambda x:unicodedata.normalize('NFKD', x).encode('ascii','ignore'))

test=pd.DataFrame.from_csv('testData.tsv', sep='\t')
test['review']=test['review'].map(lambda x: x.decode("utf8") ).\
map(lambda x:unicodedata.normalize('NFKD', x).encode('ascii','ignore'))
```

```python
def StripPunctuation(s):
    words=[]
    for word in s:
        exclude="?:!.,;.-_`~+=#()/\|][*' "
        word = ''.join(ch if ch not in exclude else " " for ch in word )
        words+=word.split()
    return words

# def Suggest(s):
#     d = enchant.Dict("en_US")
#     result=[]
#     for i,elt in enumerate(s):
#         try:
#             if not d.check(elt):
#                 result+=d.suggest(elt)[0].split()
#                 print(d.suggest(elt)[0].split(),elt)
#             else: result+=[elt]
#         except: result+=[elt]
#     return result
    

def prepareWords(review):
    stopWords = set(stopwords.words("english"))   
    s = re.sub(r'<.*/>', ' .',review)  
    s=re.sub("[^a-zA-Z]", " ", s) 
    word_list=StripPunctuation(nltk.word_tokenize(s))

            
    s=[WordNetLemmatizer().lemmatize(elt)\
       for elt in word_list if not ((elt in stopWords) or (len(elt)<2)) ]
    return( " ".join( s )) 





    
```

```python
train['review_cleaned']=train['review'].map(prepareWords)
print('train done')
test['review_cleaned']=test['review'].map(prepareWords)
```

```python
kf = cross_validation.KFold(train.shape[0], n_folds=6,shuffle=True)


train_current=train.ix[list(list(kf)[0][0])]
test_current=train.ix[list(list(kf)[0][1])]

my_pos=train_current[train_current['sentiment']==1]['review_cleaned']
my_neg=train_current[train_current['sentiment']==0]['review_cleaned']


all_pos=''
all_neg=''
for i, elt in enumerate(my_pos):
    all_pos+=" "+elt
for i, elt in enumerate(my_neg):
    all_neg+=" "+elt
    
```

```python
from sklearn.preprocessing import Normalizer

tfidfer_pos = TfidfVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 4900,\
                             lowercase=False)
X_all_pos=tfidfer_pos.fit_transform([all_pos])
X_all_pos=tfidfer_pos.fit_transform(list(my_pos))

tfidf_all_pos=pd.DataFrame(X_all_pos.toarray(),columns=tfidfer_pos.get_feature_names())

pos_train=Normalizer(copy=False).fit_transform(np.asarray(np.asmatrix(X_all_pos.toarray())))
```

```python
tfidfer_neg = TfidfVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 4900,\
                             lowercase=False)
X_all_neg=tfidfer_neg.fit_transform([all_neg])
X_all_neg=tfidfer_neg.fit_transform(list(my_neg))

tfidf_all_neg=pd.DataFrame(X_all_neg.toarray(),columns=tfidfer_neg.get_feature_names())

neg_train=Normalizer(copy=False).fit_transform(np.asarray(np.asmatrix(X_all_neg.toarray())))
```
